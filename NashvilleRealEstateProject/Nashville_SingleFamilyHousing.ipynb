```python
#remove useless or duplicate info columns
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Parcel ID','State', 'City', 'Legal Reference', 'image', 'Tax District', 'Sold As Vacant', 'Suite/ Condo   #', 'Property City', 'Multiple Parcels Involved in Sale', 'Owner Name', 'Address'], axis = 1)
```

```python
#copied from Maria Victoria's notebook

residence =['RESIDENTIAL CONDO', 'SINGLE FAMILY', 'APARTMENT: LOW RISE (BUILT SINCE 1960)','DUPLEX',
       'CONDO', 'ZERO LOT LINE', 'RESIDENTIAL COMBO/MISC', 'TRIPLEX','QUADPLEX',
                'DORMITORY/BOARDING HOUSE', 'MOBILE HOME', 'PARSONAGE']
    
commerical = ['SMALL SERVICE SHOP', 'TERMINAL/DISTRIBUTION WAREHOUSE',
       'NON-PROFIT CHARITABLE SERVICE',
       'METRO OTHER THAN OFC, SCHOOL,HOSP, OR PARK', 'NIGHTCLUB/LOUNGE',
       'MORTUARY/CEMETERY', 'PARKING LOT','OFFICE BLDG (ONE OR TWO STORIES)', 
        'CLUB/UNION HALL/LODGE', 'LIGHT MANUFACTURING',
       'ONE STORY GENERAL RETAIL STORE', 'DAY CARE CENTER','CHURCH']
    
vacant = ['VACANT RES LAND', 'VACANT COMMERCIAL LAND', 'VACANT RURAL LAND','VACANT ZONED MULTI FAMILY', 
              'VACANT RESIDENTIAL LAND','VACANT RESIENTIAL LAND']
    
other = ['GREENBELT','GREENBELT/RES\nGRRENBELT/RES']
```

```python
#create dataframe with only rows that are residence
df_housing = df[(df['Land Use'].isin(residence))]
```

```python
df_housing.head()
#df_housing.count()
```

In this dataset, I'm only going to keep the housing types that are Single Family because it has the most Neighborhood information available. I will remove all rows that don't have an address and don't have a neighborhood assigned to it.

Then, all the duplicate property addresses are removed except for the address with the most current sale date. 

The 'Neighborhood' column will be converted into a str, so it can be used as a category, and the 'Sale Price' column will be turned into a float for more precision when doing calculations using this column.

```python
#make dataframe containing only Single Family houses
df_SingleFamily = df_housing[df_housing['Land Use'] == 'SINGLE FAMILY']

df_SingleFamily.count()
```

```python
df_SingleFamily.tail()
```

```python
#delete rows w/ no property address
#delete rows with no neighborhood

df_SingleFamily = df_SingleFamily.dropna(subset=['Property Address'])
df_SingleFamily = df_SingleFamily.dropna(subset=['Neighborhood'])
df_SingleFamily.count()
```

```python
#looking at duplicate properties with multiple sale dates
address = df_SingleFamily["Property Address"]
df_SingleFamily[address.isin(address[address.duplicated()])].sort_values("Property Address")
```

```python
#change 'Sale Date' values to datetime format; assign to new column
df_SingleFamily['SaleDT'] = pd.to_datetime(df_SingleFamily['Sale Date'])

#keep most recent sale for a property address
df_SingleFamilyCurrent = df_SingleFamily.groupby('Property Address', group_keys=False, as_index=False).apply(lambda x: x.loc[np.argmax(x['SaleDT'])])
```

```python
df_SingleFamilyCurrent.count()
```

```python
df_SingleFamilyCurrent.head()
```

```python
#convert neighborhood type from float->int->str to use as categorical variable
df_SingleFamilyCurrent['Neighborhood'] = df_SingleFamilyCurrent.loc[:, ['Neighborhood']].astype(int)
df_SingleFamilyCurrent['Neighborhood'] = df_SingleFamilyCurrent.loc[:, ['Neighborhood']].astype(str)
df_SingleFamilyCurrent['Neighborhood'].dtype
```

```python
df_SingleFamilyCurrent['Sale Price'].dtype
```

```python
#convert 'Sale Price' from int->float
df_SingleFamilyCurrent['Sale Price'] = df_SingleFamilyCurrent.loc[:, ['Sale Price']].astype(float)
df_SingleFamilyCurrent['Sale Price'].dtype
```

```python
df_SingleFamilyCurrent.to_csv("Nashville_SingleFamilyHousing.csv", index=False)
```
